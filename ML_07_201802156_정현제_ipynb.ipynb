```python
def convert(imgf,labelf,outf,n):
  f = open(imgf,"rb")
  o = open(outf,"w")
  l = open(labelf,"rb")

  f.read(16)
  l.read(8)
  images = []

  for i in range(n):
    image = [ord(l.read(1))]
    for j in range(28*28):
      image.append(ord(f.read(1)))
    images.append(image)

  for image in images:
    o.write(",".join(str(pix) for pix in image)+"\n")
  
  f.close()
  o.close()
  l.close()
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import train_test_split

train_feats,test_feats,train_labels,test_labels = train_test_split(load_iris().data,load_iris().target,test_size=0.2)
```

```python
print(train_feats.shape)
print(test_feats.shape)
print(train_labels.shape)
print(test_labels.shape)
```

```python
clf = DecisionTreeClassifier()
parameters = {
    "criterion" : ["gini", "entropy", "log_loss"],
    "splitter": ["best","random"],
    "max_depth": [1,2,3,4,5],
    "random_state": [0,1,2,3,4]
}
```

```python
Gscv = GridSearchCV(clf,param_grid=parameters,scoring="accuracy")
Gscv.fit(train_feats,train_labels)
```

```python
print("best parameters: ",Gscv.best_params_)
print("best score: ",Gscv.best_score_)
```

```python
clf = Gscv.best_estimator_
print("test data에 대한 모델의 성능: ",clf.score(test_feats,test_labels))
```
