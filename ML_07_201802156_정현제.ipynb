```python
def convert(imgf,labelf,outf,n):
  f = open(imgf,"rb")
  o = open(outf,"w")
  l = open(labelf,"rb")

  f.read(16)
  l.read(8)
  images = []

  for i in range(n):
    image = [ord(l.read(1))]
    for j in range(28*28):
      image.append(ord(f.read(1)))
    images.append(image)

  for image in images:
    o.write(",".join(str(pix) for pix in image)+"\n")
  
  f.close()
  o.close()
  l.close()
```

```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("mnist_train.csv")
test_df = pd.read_csv("mnist_test.csv")
```

```python
train_df.to_csv("mnist_train_final.csv",index=False)
test_df.to_csv("mnist_test_final.csv",index=False)
```

```python
train_df
```

```python
train_a=train_df[(train_df['5']==1) | (train_df['5']==4) | (train_df['5']==6)]
```

```python
train_a
```

```python
from sklearn.model_selection import train_test_split

features = train_a.iloc[:, 1:]
labels = train_a.iloc[:, 0]

train_feats, test_feats, train_labels, test_labels = train_test_split(features, labels, test_size=0.2)
```

```python
from sklearn.svm import SVC

classifier=SVC()
parameters = {
    "kernel" : ["rbf"],
    "gamma": [0.1, 1, 5, 10],
    "C": [0.1, 1, 5, 10]
}
```

```python
from sklearn.model_selection import GridSearchCV

Gscv_new = GridSearchCV(classifier,param_grid=parameters,scoring="accuracy")
Gscv_new.fit(train_feats,train_labels)
```

```python
print("best parameters: ",Gscv_new.best_params_)
print("best score: ",Gscv_new.best_score_)
```

```python
classifier = Gscv_new.best_estimator_
print("test data에 대한 모델의 성능: ",classifier.score(test_feats,test_labels))
```
