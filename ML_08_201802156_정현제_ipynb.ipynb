```python
import pandas as pd
from functools import reduce
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/기계학습/lab8_test.csv')
```

```python
feature_key=df.iloc[:,4]
```

```python
for i in range(len(feature_key)):
  if feature_key[i] == 'Iris-virginica':
    feature_key[i] = 0
  elif feature_key[i] == 'Iris-versicolor':
    feature_key[i] = 1
  else:
    feature_key[i] = 2
```

```python
feature=feature_key.values.tolist()
```

```python
from sklearn.preprocessing import MinMaxScaler

data = df[['sepal length', 'petal length','sepal width','petal width']]

scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import adjusted_rand_score


max=0
maxnum=0;

for i in range(2,7):
  k = i
  
  model = KMeans(n_clusters = k, random_state = 10)

  model.fit(data_scale)

  df['cluster'] = model.fit_predict(data_scale)

  for i in range(k):
      plt.scatter(df.loc[df['cluster'] == i, 'sepal length'], df.loc[df['cluster'] == i, 'petal length'], 
                  label = 'cluster ' + str(i))
  
  plt.legend()
  plt.title("ARI={:5.3f}".format(adjusted_rand_score(feature, model.labels_)))
  plt.xlabel('Annual Income', size = 8)
  plt.ylabel('Spending Score', size = 8)
  plt.show()
  print("K=",k,"클러스터 레이블 확인 : \n {}".format(model.labels_))

  a = adjusted_rand_score(feature, model.labels_)
  if max < a:
    max = a
    maxnum=k
```

```python
print(max)
print(maxnum)
```
