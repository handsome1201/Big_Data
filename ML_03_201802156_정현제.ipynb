```python
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/lab3_test.csv')
```

```python
df.rename(columns={'Unnamed: 4':'종류'},inplace=True)
```

```python
plt.figure(figsize=(5, 5))

marker={"Iris-setosa":'o',"Iris-versicolor":'o',"Iris-virginica":'o'}
color={"Iris-setosa":'blue',"Iris-versicolor":'orange',"Iris-virginica":'green'}

grouped=df.groupby('종류')

for i,items in grouped:
  plt.scatter('sepal length','sepal width',data=items,s=9,label=i,marker=marker[i],c=color[i])



plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('Figure 1')
plt.legend()
plt.show()
```
